```python
from backtest import *
```

```python
import pandas as pd
import numpy as np
import datetime
```

```python
predictions = pd.read_excel('backtesting.xlsx')
```

```python
predictions = predictions.dropna()
predictions = predictions.iloc[69:,:]
predictions['date'] = predictions['date'].apply(lambda x: datetime.datetime.strptime(str(x.date()), "%Y-%m-%d"))
predictions["date"] = pd.to_datetime(predictions["date"], format='%d/%m/%Y', infer_datetime_format=True)
predictions
```

```python
# backtesting Logistic Regression and Linear Regression
```

```python
# logistic classification with sentiment
backtest_classification_noSentiment(
    list(predictions['logistic_c_with_sentiment']), 
    list(predictions['date']), 
    'logistic_c_with_sentiment_backtest')
```

```python
# logistic classification without sentiment
backtest_classification_noSentiment(
    list(predictions['logistic_c_without_sentiment']), 
    list(predictions['date']), 
    'logistic_c_without_sentiment_backtest')
```

```python
# linear regression with sentiment
backtest_regression_noSentiment(
    list(predictions['linear_r_with_sentiment']), 
    list(predictions['date']), 
    'linear_r_with_sentiment_backtest')
```

```python
# linear regression without sentiment
backtest_regression_noSentiment(list(predictions['linear_r_without_sentiment']), list(predictions['date']), 'linear_r_without_sentiment_backtest')
```
