```python
import pandas as pd
import numpy as np
```

## Backtest Results of Individual Models

```python
df = pd.read_csv('backtest_results.csv')
```

```python
for column in df.columns:
    df[column] = np.log(df[column]/df[column].shift(1))
df = df.dropna()
```

```python
output = df.describe()
output
```

```python
(output.iloc[1]/output.iloc[2]).apply(lambda x: round(x, 2))
```

```python
output.iloc[2]
```
