```python
from backtest import *
```

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
predictions = pd.read_csv('../models/ensemble_predictions.csv')
predictions = predictions.dropna()
predictions["date"] = pd.to_datetime(predictions["date"], format='%d/%m/%y', infer_datetime_format=True)
```

```python
bagging_c_without_sentiment = backtest_classification_noSentiment(
                                list(predictions['bagging_without_sentiment_classification']), 
                                list(predictions['date']), 
                                'bagging_c_without_sentiment')
bagging_c_with_sentiment = backtest_classification_WithSentiment(
                                list(predictions['bagging_with_sentiment_classification']), 
                                list(predictions['date']),
                                list(predictions['Sentiments']),
                                'bagging_c_with_sentiment')

bagging_r_without_sentiment = backtest_regression_noSentiment(
                                list(predictions['bagging_without_sentiment_regression']), 
                                list(predictions['date']), 
                                'bagging_r_without_sentiment')
bagging_r_with_sentiment = backtest_regression_WithSentiment(
                                list(predictions['bagging_with_sentiment_regression']), 
                                list(predictions['date']),
                                list(predictions['Sentiments']),
                                'bagging_r_with_sentiment')
```

```python
stacking_c_without_sentiment = backtest_classification_noSentiment(
                                list(predictions['stacking_without_sentiment_classification']), 
                                list(predictions['date']), 
                                'stacking_c_without_sentiment')
stacking_c_with_sentiment = backtest_classification_WithSentiment(
                                list(predictions['stacking_with_sentiment_classification']), 
                                list(predictions['date']),
                                list(predictions['Sentiments']),
                                'stacking_c_with_sentiment')

stacking_r_without_sentiment = backtest_regression_noSentiment(
                                list(predictions['stacking_without_sentiment_regression']), 
                                list(predictions['date']), 
                                'stacking_r_without_sentiment')
stacking_r_with_sentiment = backtest_regression_WithSentiment(
                                list(predictions['stacking_with_sentiment_regression']), 
                                list(predictions['date']),
                                list(predictions['Sentiments']),
                                'stacking_r_with_sentiment')
```

```python
res = pd.DataFrame()
res["bagging_c_without_sentiment"] = bagging_c_without_sentiment
res["bagging_c_with_sentiment"] = bagging_c_with_sentiment
res["bagging_r_without_sentiment"] = bagging_r_without_sentiment
res["bagging_r_with_sentiment"] = bagging_r_with_sentiment

res["stacking_c_without_sentiment"] = stacking_c_without_sentiment
res["stacking_c_with_sentiment"] = stacking_c_with_sentiment
res["stacking_r_without_sentiment"] = stacking_r_without_sentiment
res["stacking_r_with_sentiment"] = stacking_r_with_sentiment
```

```python
res.to_csv('ensemble_backtest_results.csv')
```
